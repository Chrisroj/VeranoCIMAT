```python
import twint

# Get tweets from different users and different words
def scrape_tweets(word, path):
    # Scraping for tweets
    c = twint.Config()
    c.Search = word
    c.Store_csv = True
    c.Output = os.path.join(path, word + "_" + "tweets" + ".csv")
    twint.run.Search(c)
    return print("Files: ", os.listdir(path))
```

# 2 Top popular users (related to BTC)

```python
# Path
path_data = r"C:\Users\ferro\Desktop\VeranoCIMAT\AnÃ¡lisis de Sentimiento\data"
bitcoin_tweets_csv = "bitcoin_tweets.csv"
```

```python
# Extract users from bitcoin_tweets.csv
chunk_list = []
#contador = 0
for chunk in pd.read_csv(os.path.join(path_data, bitcoin_tweets_csv), usecols = ["username"], chunksize = 5000, iterator = True):
    #contador += 1
    #print(contador)
    chunk_list.append(chunk[["username"]])

# Concat chunks
users = pd.concat(chunk_list, axis = 0)
users
```

```python
# Path and File
path_2 = r"C:\Users\ferro\Desktop\VeranoCIMAT\AnÃ¡lisis de Sentimiento\credentials"
file_2 = 'credentials_twitter.json'

# Load credentials
with open(os.path.join(path_2, file_2), 'r') as fp:
    credentials = json.load(fp)
consumer_key = credentials["api_consumer_key"]
consumer_secret = credentials["api_consumer_secret_key"]
access_key = credentials["api_access_token" ]
access_secret = credentials["api_access_secret_token"]

# Create authenticate object
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)

# Set authorization
auth.set_access_token(access_key, access_secret)

# Create the API object
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# Followers of a user
def followers(username, api):
    try:
        follows = api.get_user(username)._json["followers_count"]
    except:
        follows = None
    return follows
```

```python
# Get followers of each user
user_followers = []
contador = 0
for u in users["username"].unique():
    user_followers.append((u, followers(u, api)))
    contador += 1
    if contador % 100 == 0:
        print(contador, "/", len(users["username"].unique()))
```

```python
# user_followers a Data Frame
pd.DataFrame(user_followers).to_csv("user_followers.csv")
```
